```python
from tkinter import *
window = Tk()
window.geometry("600x400")
window.title("IRIS")
#validate=window.register(validate_input)

Label(window,text="Sepal_Length").grid(row=0,)
Label(window,text="Sepal_Width").grid(row=1,)
Label(window,text="Petal_Length").grid(row=2,)
Label(window,text="Petal_Width").grid(row=3,)

e1_value=StringVar()
e2_value=StringVar()
e3_value=StringVar()
e4_value=StringVar()

e1=Entry(window,font=("Helvetica", 10),textvariable=e1_value)
e2=Entry(window,font=("Helvetica", 10),textvariable=e2_value)
e3=Entry(window,font=("Helvetica", 10),textvariable=e3_value)
e4=Entry(window,font=("Helvetica", 10),textvariable=e4_value)
e5=Text(window,height=3,width=17)



e1.grid(row=0,column=1,pady=10,sticky=W)
e2.grid(row=1,column=1,pady=10,sticky=W)
e3.grid(row=2,column=1,pady=10,sticky=W)
e4.grid(row=3,column=1,pady=10,sticky=W)
e5.grid(row=4,column=1,pady=10,sticky=W)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("iris.csv")
x = df.drop("variety",axis = "columns")
y = df.variety

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25)

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)

#yp = dt.predict(x_test)


def variety():
    list1=[float(e1_value.get()),float(e2_value.get()),float(e3_value.get()),float(e4_value.get())]
    
    eg=np.array(list1)
    eg=eg.reshape(1,-1)
                   
    yp = dt.predict(eg)
   # var.set(yp)
    
    e5.insert(END,yp)
    

Button(window,text="Variety",command = variety,height=1,width=8,fg='WHITE',bg='GREY').grid(row=4,sticky=W)



window.mainloop()
```
